```python
import pandas as pd
file_path='salary.xls'
data=pd.read_excel(file_path)
print(data.head())
X=data.iloc[:,:-1]
Y=data.iloc[:,-1]
print("Features:\n",X.head())
print("Target:\n",Y.head())
import matplotlib.pyplot as plt
for column in X.columns:
  plt.figure(figsize=(8,6))
  plt.scatter(X[column],Y,alpha=0.5)
  plt.title(f'{column} vs Target')
  plt.xlabel(column)
  plt.ylabel('Target')
  plt.show()
  from sklearn.preprocessing import MinMaxScaler
  #Create a scaler object
  scaler=MinMaxScaler()
  #Normalize the features and target
  X_normalized=scaler.fit_transform(X)
  Y_normalized=scaler.fit_transform(Y.values.reshape(-1,1))
  #Convert back to DataFrame for easier handling
  X_normalized=pd.DataFrame(X_normalized, columns=X.columns)
  Y_normalized=pd.Series(Y_normalized.flatten(), name='Targer')
  #Display the normalized data
  print("Normalized Features: \n", X_normalized.head())
```
